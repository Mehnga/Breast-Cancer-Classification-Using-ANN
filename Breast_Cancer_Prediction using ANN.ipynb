```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading the dataset
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
df=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
df['label']=breast_cancer_dataset.target
```

```python
X=df.drop(columns='label',axis=1)
y=df['label']
```

```python
print(X)
```

```python
print(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scal=StandardScaler()
```

```python
X_train_std=scal.fit_transform(X_train)
X_test_std=scal.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow.keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
```

```python
s=Sequential()
s.add(Flatten(input_shape=(30,)))
s.add(Dense(units=20,activation='relu'))
s.add(Dense(units=2,activation='sigmoid'))
```

```python
s.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history=s.fit(X_train_std,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation_data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation_data'], loc = 'lower right')
```

```python
loss,accuracy=s.evaluate(X_test_std,y_test)
print(accuracy)
```
